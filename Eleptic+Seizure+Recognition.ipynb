```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras as k
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
data = pd.read_csv('D:\Python\Online Contest\Elieptic Seizure Recognition.csv')

print(data.head())
print(data.describe())
```

```python
data.isnull().any()
```

```python
data.mean(axis=0)
```

```python
train_x = data.iloc[:int(0.6*data.shape[0]),:data.shape[1]-1]
train_y = data.iloc[:int(0.6*data.shape[0]),data.shape[1]-1]


test_x = data.iloc[int(0.6*data.shape[0]):,:data.shape[1]-1]
test_y = data.iloc[int(0.6*data.shape[0]):,data.shape[1]-1]

print(train_x.head())
print(test_x.head())
```

```python
from sklearn.decomposition import PCA


pca = PCA(50)
pca.fit(train_x)
print(pca.explained_variance_ratio_)

sum(pca.explained_variance_ratio_)
```

```python
train_redX = pca.transform(train_x)


#pd.DataFrame(train_redX).head()

train_redX = pd.DataFrame(train_redX)
```

```python
#SVM

from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import validation_curve

param_range = np.logspace(-6,-1,5)
train_scores, valid_scores = validation_curve(SVC(verbose=1,kernel='rbf'),train_redX,train_y,param_name='gamma',param_range=param_range)

```

```python
print(train_scores)

print(valid_scores)
```

```python
#Adaboost

from sklearn.ensemble import AdaBoostClassifier

params = {'n_estimators':[25,50,100,200]}
clf2 = GridSearchCV(AdaBoostClassifier(),param_grid=params)

clf2.fit(train_redX,train_y)
print(clf2.score(train_redX,train_y))

```

```python
clf3 = AdaBoostClassifier(n_estimators = 50)

clf3.fit(train_redX,train_y)

print(clf3.score(train_redX,train_y))   # on increasing n_estimators model gets better for validation set

```

```python
print(clf3.score(test_x,test_y))
print(clf3.get_params)
print(clf2.score(test_x,test_y))
print(clf2.best_params_)
```

```python
#Random Forests
from sklearn.ensemble import RandomForestClassifier

params = {'n_estimators':[5,10,20,50],'max_depth':[1,5,10,25,50]}
clf4 = GridSearchCV(RandomForestClassifier(),param_grid = params)
clf4.fit(train_redX,train_y)

print(clf4.score(train_redX,train_y))

print(clf4.best_params_)   #on increasing n_estimators model gets better and generalized.
```

```python
clf3 = AdaBoostClassifier(n_estimators = 1000)
clf3.fit(train_redX,train_y)

print(clf3.score(train_redX,train_y))
```

```python
#Neural Network aka MLP
from sklearn.neural_network import MLPClassifier

clf5 = MLPClassifier(hidden_layer_sizes=(200,))
clf5.fit(train_redX,train_y)

print(clf5.score(train_redX,train_y))
print(clf5.score(test_x,test_y))                        #on increasing the hidden layer size
                                                        #model gets overfit and validation score decreases
```
